```python
#importing libraries

import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
df = pd.read_csv("geocoding1.csv")
```

# New Section

```python
#Creating an instance of Nominatim Class
geolocator = Nominatim(user_agent="my_request")

#applying the rate limiter wrapper
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=2)

```

```python
# applying geocode method to get locations

df['location_geo'] = df['Location'].apply(geocode)
```

```python
#Applying the method to pandas DataFrame to get Lat and Long

df['Lat'] = df['location_geo'].apply(lambda x: x.latitude if x else None)
df['Lon'] = df['location_geo'].apply(lambda x: x.longitude if x else None)
```

```python
# write result to csv file

df.to_csv("df_geocoded.csv")
```
