```python
#List of packages used
import osmnx as ox
import pandas as pd
import geopandas as gp
import matplotlib.pyplot as plt
import shapefile
```

```python
place = "Bangalore, India"
tags2 = {"railway" : "station"} # tags=("key":"value")
```

```python
df2 = ox.geometries_from_place(place, tags2)  #importing geometries
station = pd.DataFrame(df2) # covert to pandas
station1 = station.loc[station['operator'] == "BMRCL"] #filtering metro station data
#station1 = station1.loc[station1['disused'] != "yes"] #filtering metro station data
#station1 = station1.loc[station1['disused:railway'] != "station"] #filtering metro station data
station_sel = station1.sample(n=10) # randomly select 10 metro stations from the list of all stations
station_gdf = gp.GeoDataFrame(station_sel, crs="EPSG:4326", geometry=station_sel.geometry) #convert to geopandas
station_gdf = station_gdf.to_crs({'init': 'epsg:32643'})
station_gdf500 = station_gdf.buffer(500) # creating buffer of 500 metres
```

```python
tags4 = {"amenity" : "restaurant"}
restr = ox.geometries_from_place(place, tags4)
restr_df = pd.DataFrame(restr)
rest_gdf = gp.GeoDataFrame(restr_df, crs="EPSG:4326", geometry=restr_df.geometry)
rest_gdf = rest_gdf.to_crs({'init': 'epsg:32643'})
```

```python
myshp = open("BBMP2011.shp", "rb") #opening shapefile
mydbf = open("BBMP2011.dbf", "rb") # opening database file
bg = shapefile.Reader(shp = myshp, dbf = mydbf) # reading shapefile of bangalore
```

```python
fig, ax = plt.subplots(figsize = (10,10)) #define common axis
station_gdf500.plot(ax=ax, alpha = 0.4, color = 'red') #plot station
rest_gdf.plot(ax=ax, alpha = 0.2, color = 'blue') #plot restaurant
for shape in bg.shapeRecords(): #plot bangalore ward map
    x = [i[0] for i in shape.shape.points[:]]
    y = [i[1] for i in shape.shape.points[:]]
    bgplot = plt.plot(x,y)
plt.show() # combine all the plots
```
